![](tasks/12.png)

```python
years_column = 'Year'
names_column = 'Name'

max_year = df.groupby(years_column)[names_column].nunique().idxmax()

most_popular_name = df[df[years_column] == max_year][names_column].mode().values[0]

print(f"Найпопулярніше ім'я в {max_year} році: {most_popular_name}")
```

![](tasks/13.png)

```python
years_column = 'Year'
gender_column = 'Gender'
names_column = 'Name'

female_names_df = df[df[gender_column] == 'F']

max_year_jacob_female = female_names_df[female_names_df[names_column] == 'Jacob'].groupby(years_column).size().idxmax()

print(f"Рік, коли ім'я 'Jacob' було найпопулярнішим серед жіночих імен: {max_year_jacob_female}")
```

![](tasks/14.png)

```python
years_column = 'Year'
gender_column = 'Gender'
names_column = 'Name'

neutral_names_df = df[df.duplicated(subset=[names_column], keep=False)]

max_year_neutral_names = neutral_names_df.groupby(years_column)[names_column].nunique().idxmax()

print(f"Рік із найбільшою кількістю гендерно нейтральних імен: {max_year_neutral_names}")
```

![](tasks/18.png)

```python
years_column = 'Year'
gender_column = 'Gender'
count_column = 'Count'

gender_counts_by_year = df.groupby([years_column, gender_column])[count_column].sum().unstack()

years_with_more_girls = gender_counts_by_year[gender_counts_by_year['F'] > gender_counts_by_year['M']].index

print(f"Кількість років, коли дівчаток народжувалося більше, ніж хлопчиків: {len(years_with_more_girls)}")
```

```python
years_column = 'Year'

min_year = df[years_column].min()
max_year = df[years_column].max()

observation_years = max_year - min_year

print(f"Тривалість спостереження: {observation_years}")
```

![](tasks/23.png)

```python
years_column = 'Year'
name_column = 'Name'

name_counts_by_year = df.groupby([name_column, years_column]).size().unstack()

neutral_names = name_counts_by_year.dropna().index.tolist()

print("Найпопулярніші гендерно-нейтральні імена, присутні кожного року:")
for name in neutral_names:
    print(name)
```

![](tasks/24.png)

```python
years_column = 'Year'
name_column = 'Name'
count_column = 'Count'

name_counts_by_year = df.groupby([years_column, name_column])[count_column].sum().unstack()

popular_names_by_year = name_counts_by_year.idxmax()

most_popular_unpopular_name = popular_names_by_year.value_counts().idxmax()

print(f"Найпопулярніше серед непопулярних імен: {most_popular_unpopular_name}")
```
