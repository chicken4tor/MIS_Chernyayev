Черняєв Андрій ІПС-42 лаб4 варіант 6

\begin{equation}
  \begin{cases}
    -10.3x_{1} + 10.2x_{2} = 70 \\
    4.7x_{1} + 12.3x_{2} = 173 \\
    13.2x_{1} + 8.8x_{2} = 282
  \end{cases}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

# Функції для рівнянь
def line1(x):
    return (70 + 10.3 * x) / 10.2

def line2(x):
    return (173 - 4.7 * x) / 12.3

def line3(x):
    return (282 - 13.2 * x) / 8.8

# Діапазон значень x
x_values = np.linspace(0, 50, 100)

# Побудова графіків
plt.figure(figsize=(15, 5))

# Графік для першого рівняння
plt.subplot(1, 3, 1)
plt.plot(x_values, line1(x_values), color='blue', linestyle='dashed')
plt.title('$-10.3x_1 + 10.2x_2 = 70$')
plt.axis([0, 30, 0, 30])

# Графік для другого рівняння
plt.subplot(1, 3, 2)
plt.plot(x_values, line2(x_values), color='orange', linestyle='dotted')
plt.title('$4.7x_1 + 12.3x_2 = 173$')
plt.axis([0, 30, 0, 30])

# Графік для третього рівняння
plt.subplot(1, 3, 3)
plt.plot(x_values, line3(x_values), color='green', linestyle='dashdot')
plt.title('$13.2x_1 + 8.8x_2 = 282$')
plt.axis([0, 30, 0, 30])

plt.tight_layout()
plt.show()
```

```python
# Функції для рівнянь
def line1(x):
    return (70 + 10.3 * x) / 10.2

def line2(x):
    return (173 - 4.7 * x) / 12.3

def line3(x):
    return (282 - 13.2 * x) / 8.8

# Діапазон значень x
x_values = np.linspace(0, 50, 100)

# Побудова графіків
plt.figure(figsize=(15, 5), dpi=100)

# Графік
plt.subplot(1, 3, 1)
graphic1 = plt.plot(x_values, line1(x_values), label='line1', color='blue', linestyle='dashed')
graphic2 = plt.plot(x_values, line2(x_values), label='line2', color='red', linestyle='dotted')
graphic3 = plt.plot(x_values, line3(x_values), label='line3', color='green', linestyle='dashdot')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Графік трьох прямих')
plt.axis([0, 30, 0, 30])

plt.legend()

plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.text(15, 20, r'$-10.3x_{1} + 10.2x_{2} = 70$', color='blue', fontsize=10, rotation=49)
plt.text(0, 20, r'$13.2x_{1} + 8.8x_{2} = 282$', color='green', fontsize=10, rotation=-60)
plt.text(16, 4, r'$4.7x_{1} + 12.3x_{2} = 173$', color='red', fontsize=10, rotation=-23)

plt.fill_between([5, 10, 16], [12, 17, 8], [12.1, 10.3, 7.919], interpolate=True, color='gray', alpha=0.3)

# Коефіцієнти системи рівнянь
A = np.array([[-10.3, 10.2],
              [4.7, 12.3],
              [13.2, 8.8]])

B = np.array([70, 173, 282])

# Вирішення системи рівнянь
solution1 = np.linalg.solve(A[:2, :], B[:2])
solution2 = np.linalg.solve(np.vstack([A[0, :], A[2, :]]), np.array([B[0], B[2]]))
solution3 = np.linalg.solve(A[1:, :], B[1:])

print(solution1, " ", solution2, " ", solution3)

x1, y1 = solution1
x2, y2 = solution2
x3, y3 = solution3

ax1 = plt.scatter(x1, y1, color='red')
ax2 = plt.scatter(x2, y2, color='green')
ax3 = plt.scatter(x3, y3, color='blue')

plt.annotate('перетин line1 i line2', xy=(x1, y1), xytext=(x1-4, y1-5), arrowprops=dict(arrowstyle = '->', connectionstyle = 'arc3',facecolor='red'))
plt.annotate('перетин line1 \n i line3', xy=(x2, y2), xytext=(x1+2, y1+10), arrowprops=dict(arrowstyle = '->', connectionstyle = 'arc3',facecolor='red'))
plt.annotate('перетин line2 i line3', xy=(x3, y3), xytext=(x3, y3+4), arrowprops=dict(arrowstyle = '->', connectionstyle = 'arc3',facecolor='red'))

plt.savefig('linear_equations_plot.png', dpi=100)
plt.savefig('linear_equations_plot.jpg', dpi=100)
plt.savefig('linear_equations_plot.svg', dpi=100)
```
